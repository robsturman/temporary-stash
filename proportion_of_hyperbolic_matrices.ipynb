```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import itertools as it
%matplotlib inline 
```

```python
def compute_proportion_hyperbolic(A,B,n_max):

    hyp_prop = []
    for i in range(1,n_max):
        c = 0
        for q in it.product([A,B],repeat=i):
            w = np.eye(2)
            for j in range(len(q)):
                w = w.dot(q[j])
            #print max(np.abs(np.linalg.eigvals(w)))>1.01,np.linalg.eigvals(w)
            if max(np.abs(np.linalg.eigvals(w)))>1.001:
                c=c+1
        #print i,c,float(c)/(2**i)
        hyp_prop.append(float(c)/(2**i))
    return hyp_prop
```

The obviously all-hyperbolic case.

```python
A = np.array([[2,1],[1,1]])
B = np.array([[5,2],[2,1]])
hyp_data = compute_proportion_hyperbolic(A,B,15)
plt.plot(range(1,15),hyp_data,'ro-')
plt.show()
```

The almost-obvious two positive shears case - here all products are hyperbolic except all $A$'s or all $B$'s. The proportion of hyperbolic matrices grows as $1-2^{1-k}$.

```python
A = np.array([[1,1],[0,1]])
B = np.array([[1,0],[1,1]])
hyp_data = compute_proportion_hyperbolic(A,B,15)
plt.plot(range(1,15),hyp_data,'ro')
plt.plot(range(1,15),[1.0-2.0**(1-k) for k in range(1,15)],'g-')
plt.show()
```

The far-from obvious $\sigma_1 \sigma_2$ case.

```python
A = np.array([[1,1],[0,1]])
B = np.array([[1,0],[-1,1]])
hyp_data = compute_proportion_hyperbolic(A,B,23)
plt.plot(range(1,23),hyp_data,'ro-')
plt.show()
```

The two protocols with equal homological topological entropy, $\sigma_1 \sigma_2^{-1}$ and $\sigma_1 \sigma_2^5$ are also always hyperbolic.

```python
A = np.array([[2,1],[1,1]])
B = np.array([[-4,-5],[1,1]])
hyp_data = compute_proportion_hyperbolic(A,B,15)
plt.plot(range(1,15),hyp_data,'ro-')
plt.show()
```

Two other non-trivial cases:

```python
A = np.array([[1,2],[0,1]])
B = np.array([[1,0],[-1,1]])
hyp_data = compute_proportion_hyperbolic(A,B,16)
plt.plot(range(1,16),hyp_data,'ro-')
plt.show()
```

This one goes roughly like $1-4^{1-k}$:

```python
A = np.array([[1,2],[0,1]])
B = np.array([[1,0],[-2,1]])
hyp_data = compute_proportion_hyperbolic(A,B,13)
plt.plot(range(1,13),hyp_data,'ro-')
plt.plot(range(1,15),[1.0-4.0**(1-(k+1)/2) for k in range(1,15)],'g-')
plt.ylim([0.0,1.0])
plt.show()
```
